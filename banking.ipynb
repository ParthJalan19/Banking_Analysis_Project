```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as mp
```

```python
df = pd.read_csv('Banking Analysis-20251211T134648Z-3-001\Banking Analysis\Banking.csv')
df.head(5)
```

```python
df.shape
```

```python
#generative descriptive statistics for dataframe 
df.describe()
```

```python
df['Estimated Income'].min()
```

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'],bins = bins, labels = labels, right = False)
```

```python
df['Income Band'].value_counts().plot(kind='bar') 
```

```python
#Examine the distribution of unique categories in categorical columns 
categorical_cols = [
    "BRId",
    "GenderId",
    "IAId",
    "Amount of Credit Cards",
    "Nationality",
    "Occupation",
    "Fee Structure",
    "Loyalty Classification",
    "Properties Owned",
    "Risk Weighting"
]

# Display value counts for each categorical column
for col in categorical_cols:
    print(f"\nValue Counts for '{col}':")
    display(df[col].value_counts())
```

```python
predictors = [
    "BRId",
    "GenderId",
    "IAId",
    "Amount of Credit Cards",
    "Nationality",
    "Occupation",
    "Fee Structure",
    "Loyalty Classification",
    "Properties Owned",
    "Risk Weighting"
]

for i, predictor in enumerate(predictors):
    plt.figure(i, figsize=(10,4))
    sns.countplot(data=df, x=predictor, hue="GenderId")
    plt.title(f"Countplot of {predictor} by Gender")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
predictors = [
    "BRId",
    "GenderId",
    "IAId",
    "Amount of Credit Cards",
    "Nationality",
    "Occupation",
    "Fee Structure",
    "Loyalty Classification",
    "Properties Owned",
    "Risk Weighting"
]

for i, predictor in enumerate(predictors):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue ='Nationality')
```

```python
#Histroplot of values count for different Occupation

for col in categorical_cols:
    if col == "Occupation":
        continue
plt.figure(figsize=(8,4))
sns.histplot(df[col])
plt.title('Histrogram of Occupation Count')
plt.xlabel(col)
plt.ylabel("Count")
plt.show()
```

```python
# numerical_cols = ['Estimated Income', 'Superannuation Savings']

#Univariate analysis and visualization

numerical_cols = ['Estimated Income', 'Superannuation Savings']

plt.figure(figsize=(10, 6))

for i, col in enumerate(numerical_cols):
    plt.subplot(2, 1, i + 1)   # 2 rows, 1 column
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
numerical_cols = [
    'Estimated Income',
    'Superannuation Savings',
    'Credit Card Balance',
    'Bank Loans',
    'Bank Deposits',
    'Checking Accounts',
    'Saving Accounts',
    'Foreign Currency Account',
    'Business Lending'
]

# Compute correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Plot heatmap
plt.figure(figsize=(12, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Heat Correlation Matrix")
plt.show()
```
